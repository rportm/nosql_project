# Storing Image in mongoDB

```python
from pymongo import MongoClient
import base64
import requests

# Connect to MongoDB (adjust the connection string as necessary)
client = MongoClient('mongodb://localhost:27017/')
db = client['card_database']
collection = db['cards']

# Fetch the image
card_name = "Dark Magician"
response = requests.get(f"https://db.ygoprodeck.com/api/v7/cardinfo.php?name={card_name}")
card_data = response.json()
image_url = card_data['data'][0]['card_images'][0]['image_url']
image_response = requests.get(image_url)

# Encode the image to base64
image_b64 = base64.b64encode(image_response.content)

# Save the encoded image to MongoDB
collection.insert_one({
    'name': card_name,
    'image': image_b64
})

# Retrieve the image
card_document = collection.find_one({'name': card_name})

# Decode the image (for demonstration, showing how to reverse the process)
image_decoded = base64.b64decode(card_document['image'])

# To display the image in a Jupyter Notebook after retrieval:
from IPython.display import display, Image
display(Image(data=image_decoded))
```
